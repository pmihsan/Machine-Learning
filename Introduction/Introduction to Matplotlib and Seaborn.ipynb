# Matplotlib

## Import

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## Line plot

```python
x = np.array([0, 6, 5, 9])
y = np.array([0, 80, 20, 90])
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, linestyle="dotted")
plt.show()
```

```python
plt.plot(x, y, linestyle="dashed")
plt.show()
```

## Bar Chart

```python
x = np.array(["X", "Y", "Z", "M"])
y = np.array([20, 30, 25, 22])
```

```python
plt.bar(x, y)
plt.show()
```

## Scatter Plot

```python
x1 = np.array([20, 15, 11, 7, 8, 10])
y1 = np.array([2, 7, 5, 12, 8, 10])
```

```python
plt.scatter(x1, y1)
plt.show()
```

## Pie chart

```python
 p = np.array([20, 30, 40, 12])
 plt.pie(p)
```

## Historgram

```python
h1 = np.random.normal(170, 10, 250)
print(h1)
```

```python
plt.hist(h1)
```

# Seaborn


## Import

```python
import seaborn as sns
import pandas as pd
```

## Box Plot

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.boxplot(by = 'day', column = ['total_bill'], grid=False)
```

## Dist Plot


```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
age1 = titanic['age'].dropna()
sns.distplot(age1, bins=30, kde=False)
plt.show()
```

## Reg Plot

```python
data = sns.load_dataset("mpg")
```

```python
data.head()
```

```python
sns.regplot(x="mpg", y="acceleration", data=data)
plt.show()
```
