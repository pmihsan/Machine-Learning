## Series

```python
a = [1 ,7, 2]
var = pd.Series(a)
print(var)
```

## Dataframe

```python
data = {
    "calories": [230, 420, 390],
    "duration": [45, 56, 80]
}

data = pd.DataFrame(data)
print(data)
```
