## Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

## Preprocessing


```python
data = pd.read_csv("./diabetes.csv")
```

```python
data.describe()
```

```python
data.info()
```

```python
x = data.drop(['Outcome'], axis=1)
y = data['Outcome']
```

```python
x
```

```python
y
```

```python
x.isnull().sum()
```

## Train and Test Sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# x_train, x_test, y_train, y_test
y_train
```

## Scaling Training and Testing Sets

```python
scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train,x_test
```

## MLP Classifier

```python
classfier = MLPClassifier(hidden_layer_sizes=(30,30,30))
classfier.fit(x_train, y_train)
```

```python
predictions = classfier.predict(x_test)
predictions
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
accuracy_score(y_test,predictions)
```

## Comparing With Neural Network

```python
nnModel = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1, max_iter = 1000)
```

```python
nnModel.fit(x,y)
```

```python
print (f'Accuracy - : {nnModel.score(x,y):.3f}')
```
